```python
import pandas as pd
import matplotlib.pyplot as plt

# === Read CSV ===
df = pd.read_csv("metrics_summary.csv")

# Take first 5 runs (as you want)
df0 = df.head(5)

runs = range(1, len(df0) + 1)

# === Plot 1: Bytes per Report ===
plt.figure()
plt.plot(runs, df0["bytes_per_report"], marker='o')
plt.xlabel("Run")
plt.ylabel("Bytes per Report")
plt.title("Bytes per Report vs Run")
plt.grid(True)
plt.show()

# === Plot 2: Packets Received ===
plt.figure()
plt.plot(runs, df0["packets_received"], marker='o')
plt.xlabel("Run")
plt.ylabel("Packets Received")
plt.title("Packets Received vs Run")
plt.grid(True)
plt.show()

# === Plot 3: Duplicate Rate ===
plt.figure()
plt.plot(runs, df0["duplicate_rate"], marker='o')
plt.xlabel("Run")
plt.ylabel("Duplicate Rate")
plt.title("Duplicate Rate vs Run")
plt.grid(True)
plt.show()

# === Plot 4: Sequence Gap Count ===
plt.figure()
plt.plot(runs, df0["sequence_gap_count"], marker='o')
plt.xlabel("Run")
plt.ylabel("Sequence Gap Count")
plt.title("Sequence Gap Count vs Run")
plt.grid(True)
plt.show()

# === Plot 5: CPU ms per Report ===
plt.figure()
plt.plot(runs, df0["cpu_ms_per_report"], marker='o')
plt.xlabel("Run")
plt.ylabel("CPU ms per Report")
plt.title("CPU Cost per Report vs Run")
plt.grid(True)
plt.show()
```
